```python
a=[]
for i in range(869601,874601):
    a.append(i)
```

```python

url_punjabi="http://jagbani.punjabkesari.in/life-style/news/salsa-chicken-"
strs =[]
for j in ((a)):
    strs.append(url_punjabi+""+str(j))
```

```python
list_arr=[]
list_headding=[]

for i in(strs[3000:4000]):
    print(i)
    r  = requests.get(i)
    data = r.text
    soup = BeautifulSoup(data,'html5lib')
    
    for headding in (strs[3000:4000]):
        important_headding=soup('h2',{'class':'ttlls'})
    list_headding.append(important_headding)
    
    for j in (strs[3000:4000]):
        all_paragraphs=soup.find_all('p')
    list_arr.append(all_paragraphs)
  
    
    
```

```python
import pickle
with open("list_arr.pickle","wb") as p:
    pickle.dump(list_arr,p)
```

```python
import re
clean_list=[]
for i in range(0,len(list_arr)):
    punjabi_string=str(tuple(map(str,list_arr[i])))
    #print(punjabi_string)
    docu_string=re.sub("[\n\" \n, '\n'  & , /  _ < > % $ * # @ \\xa0  = // ; \\   ....)'(\  \-!?:|]+",' ',punjabi_string)
    document_string=docu_string.replace('\\n', '').replace('\\xa0','').replace('\\','').replace('px','').split()
    clean_list.append(document_string)

    
clean_words=[]
for i in range(0,len(clean_list)):
    clean=[]
    for word in clean_list[i]:
        if not word.isalpha():
            clean.append(word)
    clean_words.append(clean)
    
    
articles=[]
for i in range(0,len(clean_words)):
    result = " ".join(clean_words[i])
    articles.append(result)
articles
```

```python
import re
clean_list=[]
for i in range(0,len(list_headding)):
    punjabi_string=str(tuple(map(str,list_headding[i])))
    #print(punjabi_string)
    docu_string=re.sub("[\n\" \n, '\n'  & , /   < >   \\xa0  = // \\   ....)'(\  \-!?:|]+",' ',punjabi_string)
    document_string=docu_string.replace('\\n', '').replace('\\xa0','').replace('\\','').split()
    clean_list.append(document_string)

    
clean_words=[]
for i in range(0,len(clean_list)):
    clean=[]
    for word in clean_list[i]:
        if not word.isalnum():
            clean.append(word)
    clean_words.append(clean)
    
    
headdings=[]
for i in range(0,len(clean_words)):
    result = " ".join(clean_words[i])
    headdings.append(result)
headdings[0]
```
